```python
# 6.1 Membuat Kelas Node dan Linked List

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
# 6.2 Implementasi Singly Linked List

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
```

```python
# 6.3 Implementasi Doubly Linked List

class DoublyNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = DoublyNode(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
```

```python
# 6.4 Implementasi Circular Linked List

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head

    def display(self):
        if not self.head:
             return
        temp = self.head
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.head:
                break
        print('(Back to Head)')
```

```python
# Definisi Node dalam Linked List
class TaskNode:
    def __init__(self, task_name):
        self.task_name = task_name # Menyimpan nama tugas
        self.next = None # Pointer ke tugas berikutnya

# Definisi Linked List untuk Daftar Tugas
class TaskList:
    def __init__(self):
        self.head = None # Awalnya tidak ada tugas (daftar kosong)

    # Menambahkan tugas ke Linked List
    def add_task(self, task_name):
        new_task = TaskNode(task_name)
        if not self.head:
            self.head = new_task # Jika list kosong, tugas pertama menjadi Head
        else:
            current = self.head
            while current.next: # Mencari node terakhir dalam list
                current = current.next
            current.next = new_task # Menambahkan tugas di akhir list

    # Menampilkan daftar tugas
    def show_tasks(self):
        if not self.head:
            print("\n Tidak ada tugas hari ini!")
            return

        print("\n Daftar Tugas Hari Ini:")
        current = self.head
        count = 1
        while current:
            print(f"{count}. {current.task_name}")
            current = current.next
            count += 1

# ====== Program Utama (Main) ====== #

# Membuat objek daftar tugas
task_list = TaskList()

# Meminta pengguna memasukkan jumlah tugas
num_tasks = int(input("Masukkan jumlah tugas hari ini: "))

# Loop untuk memasukkan tugas ke dalam Linked List
for i in range(1, num_tasks + 1):
    task_name = input(f"Masukkan tugas ke-{i}: ")
    task_list.add_task(task_name)

# Menampilkan hasil daftar tugas
task_list.show_tasks()
```

2. Implementasi Doubly Linked List


```python
# Definisi Node dalam Doubly Linked List
class PageNode:
    def __init__(self, url):
        self.url = url # URL halaman web
        self.next = None # Pointer ke halaman berikutnya
        self.prev = None # Pointer ke halaman sebelumnya

# Definisi Doubly Linked List untuk Riwayat Peramban
class BrowserHistory:
    def __init__(self):
        self.current = None # Pointer ke halaman yang sedang dibuka

    # Menambahkan halaman baru ke dalam riwayat
    def visit_page(self, url):
        new_page = PageNode(url)
        if not self.current:
            self.current = new_page # Jika riwayat kosong, halaman ini menjadi halaman pertama
        else:
            new_page.prev = self.current
            self.current = new_page # Pindahkan posisi ke halaman baru

    # Menampilkan riwayat kunjungan
    def show_history(self):
        if not self.current:
            print("\n Riwayat kosong.")
            return

        print("\n Riwayat Peramban web:")
        temp = self.current
        while temp.prev: # Kembali ke halaman pertama
            temp = temp.prev

        count = 1
        while temp:
            print(f"{count}. {temp.url}")
            temp = temp.next
            count += 1

    # Navigasi mundur ke halaman sebelumnya
    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"\n Anda kembali ke: {self.current.url}")
        else:
            print("\n Tidak bisa kembali, ini adalah halaman pertama!")

    # Navigasi maju ke halaman berikutnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"\n Anda maju ke: {self.current.url}")
        else:
            print("\n Tidak bisa maju, ini adalah halaman terakhir!")

# ====== Program Utama (Main) ====== #

# Membuat objek riwayat peramban
history = BrowserHistory()

# Meminta pengguna memasukkan jumlah halaman yang dikunjungi
num_pages = int(input("\nMasukkan jumlah halaman yang dikunjungi: "))

# Loop untuk memasukkan URL halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
    url = input(f"Masukkan URL halaman ke-{i}: ")
    history.visit_page(url)

# Menampilkan riwayat yang telah dimasukan
history.show_history()

# Menu navigasi pengguna
while True:
    command = input("\n Tekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower ()

    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward()
    elif command == 'x':
        print("\n Keluar dari riwayat peramban.")
        break
    else:
        print("\n Perintah tidak dikenal.")
```

LATIHAN 2 : Buatlah program yang menampilkan riwayat navigasi folder dalam File Explorer menggunakan Doubly Linked List. Jumlah folder yang Anda navigasikan sesuai dengan digit terakhir NPM Anda!

```python
#Latihan 2
class PageNode:
  def __init__(self, folder):
    self.folder = folder
    self.next = None
    self.prev = None

class FolderHistory:
  def __init__(self):
    self.current = None

  def visit_folder(self, folder):
      new_folder = PageNode(folder)
      if not self.current:
        self.current = new_folder

      else:
        new_folder.prev = self.current
        self.current.next = new_folder
        self.current = new_folder

  def show_history(self):
      if not self.current:
        print("\n File history is empty!")
        return

      print("\n File History:")
      temp = self.current
      while temp.prev:
        temp = temp.prev

      count = 1
      while temp:
        print(f"{count}. {temp.folder}")
        temp = temp.next
        count += 1

  def go_back(self):
      if self.current and self.current.prev:
        self.current = self.current.prev
        print(f"\n Anda kembali ke: {self.current.folder}")

      else:
        print("\n Tidak ada halaman sebelumnya!")

  def go_forward(self):
      if self.current and self.current.next:
        self.current = self.current.next
        print(f"\n Anda maju ke: {self.current.folder}")

      else:
        print("\n Tidak ada halaman berikutnya!")

history = FolderHistory()

npm = int(input("NPM: "))
jumlah = npm % 10

if jumlah == 0:
  print("Tidak ada.")
else:
  history = FolderHistory()

for i in range (1, (npm % 10) + 1):
  folder = input(f"Masukkan folder halaman ke-{i}:  ")
  history.visit_folder(folder)

history.show_history()

while True:
  command = input("\n Tekan 'b' untuk kembali, 'f' untuk maju, tekan 'x' untuk keluar")

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari riwayat folder.")
    break
  else:
    print("\n Perintah tidak dikenal.")
```

3. Implementasi Circular Linked List

```python
# Definisi Node dalam Circular Linked List
class SongNode:
    def __init__(self, title):
        self.title = title # Menyimpan judul lagu
        self.next = None # Pointer ke lagu berikutnya

# Definisi Circular Linked List untuk Playlist Lagu
class Playlist:
    def __init__(self):
        self.current = None # Lagu yang sudah diputar

    # Menambahkan lagu ke dalam playlist
    def add_song(self, title):
        new_song = SongNode(title)
        if not self.current:
            self.current = new_song
            self.current.next = self.current # Menghubungkan ke dirinya sendiri (Circular)
        else:
            temp = self.current
            while temp.next != self.current: # Mencari lagu terakhir
                temp = temp.next
            temp.next = new_song # Menambahkan lagu di akhir
            new_song.next = self.current # Menghubungkan kembali ke awal

    # Menampilkan daftar lagu dalam playlist
    def show_playlist(self):
        if not self.current:
            print("\n Playlist kosong!")
            return

        print("\n Playlist Lagu Anda:")
        temp = self.current
        count = 1
        while True:
            print(f"{count}. {temp.title}")
            temp = temp.next
            count += 1
            if temp == self.current:  # Berhenti jika sudah kembali ke awal
                break

    # Memutar lagu berikutnya dalam playlist
    def next_song(self):
        if self.current:
            self.current = self.current.next
            print(f"\n Memutar: {self.current.title}")

# ====== Program Utama (Main) ====== #

# Membuat objek playlist lagu
playlist = Playlist()

# Meminta pengguna memasukkan jumlah lagu
num_songs = int(input("\nMasukkan jumlah lagu dalam playlist: "))

# Loop untuk memasukkan lagu ke dalam Circular Linked List
for i in range(1, num_songs + 1):
    title = input(f"Masukkan judul lagu ke-{i}: ")
    playlist.add_song(title)

# Menampilkan daftar lagu
playlist.show_playlist()

# Menu navigasi pengguna
while True:
    command = input("\n Tekan 'n' untuk memutar lagu berikutnya, Tekan 'x' untuk keluar.\nMasukkan perintah: ").lower()

    if command == 'n':
        playlist.next_song()
    elif command == 'x':
        print("\n Keluar dari pemutar musik.")
        break
    else:
        print("\n Perintah tidak dikenal.")
```

4. Operasi Dasar pada Linked List

```python
# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data):
        if position < 0:
            print("Posisi tidak valid.")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for i in range(position - 1):
            if not temp:
                print("Posisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("Daftar kosong. Tidak ada yang dapat dihapus!")
            return
        self.head = self.head.next

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong. Tidak ada yang dapat dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position):
        if not self.head:
            print("Daftar kosong. Tidak ada yang dapat dihapus!")
            return
        if position == 0:
            self.head = self.head.next
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp or not temp.next:
                print("Posisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f" Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print(f" Elemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        if not self.head:
            print("Daftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

# ====== Programm Utama (Main) ====== #

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("Masukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
    data = int(input(f"Masukkan data ke-(i+1): "))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\n Pilih Operasi:")
    print("1. Tambah Elemen di awal")
    print("2. Tambah Elemen di akhir")
    print("3. Tambah Elemen di posisi Tertentu")
    print("4. Hapus Elemen dari awal")
    print("5. Hapus Elemen dari akhir")
    print("6. Hapus Elemen dari posisi Tertentu")
    print("7. Cari Elemen dalam Daftar")
    print("8. Balikkan Daftar")
    print("9. Tampilkan Daftar")
    print("10. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print(" Daftar berhasil dibalik")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\n Keluar dari program.")
        break
    else:
        print("\n Pilihan tidak valid, silakan coba lagi!")
```
